```python
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]
```

```python
# 1st Method
```

```python
fig = plt.figure() # creates a figure

ax = fig.add_subplot() # add some axes

plt.show();
```

```python
fig = plt.figure() # create a figure

ax = fig.add_axes([1, 1, 1, 1])

ax.plot(x, y) # add some data

plt.show()
```

```python
# 3rd method (recommended)
```

```python
fig, ax = plt.subplots()

ax.plot(x, y); # add some data
```

```python
fig, ax = plt.subplots()

ax.plot([50, 100, 200, 250]); # add some data
```

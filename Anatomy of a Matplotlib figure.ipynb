# Matplotlib Example Workflow

```python
# 0. import matplotlib and get it ready for our notebook

%matplotlib inline
import matplotlib.pyplot as plt


# 1. Prepare Data



x = [1, 2, 3, 4]

y = [11, 22, 33, 44]



# 2. Set up plot



fig, ax = plt.subplots(figsize=(10, 10))



# 3. Plot Data



ax.plot(x, y);



# 4. Customise Plot



ax.set(title='Simple Plot',
      xlabel='x-axis',
      ylabel='y-axis');



# 5. Save and show



fig.savefig('sample-plot.png')

```

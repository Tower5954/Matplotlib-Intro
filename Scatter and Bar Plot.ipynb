```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

# Making Figures with NumPy arrays

We want:

* Line plot
* Scatter plot
* Bar plot
* Histogram
* Subplots

```python
# Create some Data

x = np.linspace(0, 10, 100)
x
```

```python
x[:10]
```

```python
# plot the data and create a line plot

fig, ax = plt.subplots()
ax.plot(x, x**2);
```

```python
fig, ax = plt.subplots()
ax.plot(x, x**3);
```

```python
fig, ax = plt.subplots()
ax.plot(x, x**7);
```

```python
# Use same data to make a scatterplot
```

```python
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x));
```

```python
# Another scatter plot
```

```python
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x));
```

```python
# Make a plot from a dictionary
```

```python
car_prices = {'Tesla': 10000,
              'Honda': 7000,
              'Mercedes': 8000,}
```

```python
fig, ax = plt.subplots()
ax.bar(car_prices.keys(),
       car_prices.values());
```

```python
fig, ax = plt.subplots()
ax.bar(car_prices.keys(),
       car_prices.values())
ax.set(title='Cars', 
       ylabel='Prices');
```
